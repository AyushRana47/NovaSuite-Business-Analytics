###  View Column Names



```python


print('customers columns :', list(customers.columns))
print('plans     columns :', list(plans.columns))
print('revenue   columns :', list(revenue.columns))
print('usage     columns :', list(usage.columns))
print('churn     columns :', list(churn.columns))
```

```python

col_customers = list(customers.columns)
col_plans     = list(plans.columns)
col_revenue   = list(revenue.columns)
col_usage     = list(usage.columns)
col_churn     = list(churn.columns)

# Combine all column names into one list with their table name
all_columns = []
for col in col_customers:
    all_columns.append(('customers', col))
for col in col_plans:
    all_columns.append(('plans', col))
for col in col_revenue:
    all_columns.append(('revenue', col))
for col in col_usage:
    all_columns.append(('usage', col))
for col in col_churn:
    all_columns.append(('churn', col))

# Count how many tables each column name appears in
col_count = {}
for table, col in all_columns:
    if col not in col_count:
        col_count[col] = []
    col_count[col].append(table)

print('Columns that appear in more than one table (potential join keys):')
print('-' * 55)
for col, tables in col_count.items():
    if len(tables) > 1:
        print(f'  {col:25s} → found in: {tables}')
```
